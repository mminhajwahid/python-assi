```python
# Q4) python program to sum all the numeric item in dictionary
d={'A':10,'B':50,'C':29,'d':'ddd'}
lst = []
for x in d.values():
    if type(x)==int or type(x) == float:
        lst.append(x)
print(sum(lst))
```

```python
# Q5) program to identify duplicate values from list
l = [1,2,3,4,1,2,'as','dfg','as']
l1 = []
for i in l:
    if i not in l1:
        l1.append(i)
    else:
        print(i, end=' ')
```

```python
# Q6) python script to check if a given key already exisits in a dictionary
def checkKey(dict, key):
      
    if key in dict.keys():
        print("Present, ", end =" ")
        print("value =", dict[key])
    else:
        print("Not present")
  
# Driver Code
dict = {'a': 100, 'b':200, 'c':300}
  
key = 'b'
checkKey(dict, key)
  
key = 'w'
checkKey(dict, key)
```
